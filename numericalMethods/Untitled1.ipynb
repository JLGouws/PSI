# Ordinary Differential Equations

```julia
using DifferentialEquations
```

```julia
] add DifferentialEquations
```

```julia
function RK(x_0, y_0, f, h, N)
    for i in 1:N
        ys = [y_0]
        k_0 = f(x_0, y_0)
        y_1 = y_0 + h * k_0 / 2
        x_1 = x_0 + h / 2
        k_1 = f(x_1, y_1)
        y_2 = y_0 + h  * k_1
        x_0 += h
        push!(ys, y_2)
    end
    
end
```

## Harmonic Oscillator

Conditions
$$ x, v $$

$$ \dot{x} = v $$
$$ \dot{v} = -x$$

```julia
function fho(U::Vector)
    x, v = U
    xdot = v
    vdot = -x
    [xdot, vdot]
end
```

```julia
fho([1, 0])
```

```julia
function eulerStep(f, y0, h)
    k0 = f(y0)
    y1 = y0 + h * k0
    y1
end
```

```julia
function eulerMethod(f, U0, h, N)
    Us = []
    U = U0
    for i in 1:N
        U = eulerStep(f, U, h)
        push!(Us, U)
    end
    Us
end
```

```julia
eulerMethod(fho, [1.,0.], 0.001, 1000)
```
